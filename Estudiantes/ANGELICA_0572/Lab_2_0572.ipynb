# Solución

Primero vamos a hacer el histograma donde se la añade el argumento "kde" y luego el fit usando "sns.relplot" con el argumento "ci".

```
# Gráfico de histograma con KDE
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(data=df_mpg, x='mpg', ax=ax[0], color="skyblue")
ax[0].set_title("Distribución MPG")
sns.histplot(data=df_mpg, x='horsepower', ax=ax[1], color = "orange")
ax[1].set_title("Distribución Horsepower");
```

```
fig, ax = plt.subplots(figsize=(10, 5))
sns.regplot(
    data=df_mpg,
    x="horsepower",
    y="mpg",
    ci=95,
    scatter_kws={"color": "orange"},
    line_kws={"color": "darkviolet"}
)

plt.title("MPG vs Horsepower (95% CI)")
plt.show()
```
